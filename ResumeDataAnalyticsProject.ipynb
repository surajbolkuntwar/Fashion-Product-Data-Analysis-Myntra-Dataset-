####  Brand Analysis 

```python
brand_name_analysis = df.groupby(['brand_name']).agg({
                    'price': ['mean', 'median', 'count'],
                    'MRP': 'mean',
                    'number_of_ratings': 'sum',
                    'ratings': 'mean',
                    'discount_percent': 'mean'})
brand_name_analysis.columns = ['avg_price', 'median_price', 'sales_count',
                          'avg_mrp',
                          'sum_no_of_ratings',
                          'avg_rating',
                          'avg_discount']

brand_name_analysis_sorted = brand_name_analysis.sort_values(by='sales_count', ascending=False)
brand_name_analysis_sorted.head(10)
```

#### Filter brand that appears at least 250 times 

```python
# For filtered brand which has appear at least 250 times sales_count

Total_MRP =df["MRP"].sum()
print("Total MRP of all the brands in dataset is:", Total_MRP,"\n")

Total_price =df["price"].sum()
print("Total price of all the brands in dataset is:", Total_price,"\n")


# Get brands that appear at least 250 times
popular_brands = df["brand_name"].value_counts()[df["brand_name"].value_counts() >=250].index

# Filter the original DataFrame based on those popular brands
df_filtered = df[df["brand_name"].isin(popular_brands)]

Total_MRP_new =df_filtered["MRP"].sum()
print("Total MRP of top 250  brands in dataset is:", Total_MRP_new,"\n")

Total_price_new =df_filtered["price"].sum()
print("Total price of top 250 brands in dataset is:", Total_price_new,"\n")

# Filter brand analysis

filter_brand_analysis = df_filtered.groupby(['brand_name']).agg({
                    'price': ['mean', 'median', 'count'],
                    'MRP': 'mean',
                    'number_of_ratings': 'sum',
                    'ratings': 'mean',
                    'discount_percent': 'mean'})
filter_brand_analysis.columns = ['avg_price', 'median_price', 'sales_count',
                          'avg_mrp',
                          'sum_no_of_ratings',
                          'avg_rating',
                          'avg_discount']

filter_brand_analysis_sorted =filter_brand_analysis.sort_values(by='sales_count', ascending=False)
filter_brand_analysis_sorted.head(10)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(7, 15)) 

# Top brand Analysis 
top_brands = df['brand_name'].value_counts().head(10)
axes[0].bar(top_brands.index, top_brands.values, color='#580F41')
axes[0].set_title('Top Brands by Sales Count', fontweight='bold')
axes[0].set_xlabel("Brand Name")
axes[0].set_ylabel("Sales Count")
axes[0].tick_params(axis='x', rotation=90)

# Top brand by ratings
brand_rating = df.groupby('brand_name').ratings.agg(['mean']).sort_values(by='mean',ascending=False)
top_brand_rating = brand_rating.head(10)

sns.scatterplot(x=brand_rating.head(10).index,y=top_brand_rating["mean"],ax=axes[1])
axes[1].set_title('Scatter Plot of brand name,avg rating',fontweight='bold')
axes[1].set_xlabel("Brand Name")
axes[1].set_ylabel("avg rating")
axes[1].tick_params(axis='x', rotation=90)

# Top brand on avg. Rating vs avg. Price
top_brand_rvp = df.groupby('brand_name').agg({'ratings': 'mean', 'price': 'mean'}).reset_index().sort_values(by='ratings', ascending=False).head(10)
sns.barplot(x=top_brand_rvp['brand_name'], y=top_brand_rvp['price'], hue=top_brand_rvp["ratings"],ax=axes[2])
plt.xticks(rotation=90)
plt.xlabel("Brand Name")
plt.ylabel("Average Prices")
plt.title("Brand avg. Prices with avg. Ratings")
plt.legend(title="Average Ratings")

fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Myntra E-commerce Analytics Dashboard', fontsize=16, fontweight='bold')

# Scatter Plot of discount_percent vs rating
sns.scatterplot(x=df["discount_percent"],y=df["ratings"],ax=axes[0,0],color="#Ed4552")
axes[0,0].set_title('Scatter Plot of discount_percent,rating')

# Scatter Plot of price vs rating
sns.scatterplot(x=df["price"],y=df["ratings"],ax=axes[0,1],color="#D5A330")
axes[0,1].set_title('Scatter Plot of price,rating')


# Create price category to analysis 
df['price_category'] = pd.cut(df['price'],
                                 bins = [0, 500, 1000, 2000, 5000, 10000, 20000, df['price'].max()],
                                labels = ['<500', '500-1k', '1k-2k', '2k-5k', '5k-10k', '10k-20k', '20k+'])

# Discount percent vs price Category
sns.barplot(data=df, x='price_category', y='discount_percent', ax=axes[1,0], color="#C96160")
axes[1,0].set_title('Bar plot on Discount percent by Price Category', fontweight='bold')
axes[1,0].tick_params(axis='x', rotation=45)

# Price category vs Rating 
df['price_category'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Price Range')
plt.ylabel('Number of Pants')
axes[1,1].set_title('Price range Distribution', fontweight='bold')
axes[1,1].tick_params(axis='x', rotation=45)


plt.tight_layout()
plt.show()
```
